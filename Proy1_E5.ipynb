## 1. Factorial de N ‚ú®

> **Datos de entrada**: `n`, n√∫mero que se calcular√° su factorial, con $n\geq0$.  
**Datos de salida**: `fact`, factorial de `n`.

El siguiente c√≥digo, define la funci√≥n `factorial` que calcula el factorial de un n√∫mero entero $n$ dado. El factorial de un n√∫mero entero positivo $n$ se define como el producto de todos los n√∫meros enteros positivos desde $n$ hasta $1$. Considerando que $0! =1$, es decir: $n! = n \times (n-1) \times(n-2)\times \cdots \times 1$. Recursivamente, esto es: 
$n! = \begin{cases} n\times (n-1)! & \text{ si } n>0 \\ 1 & \text{ si } n=0 \end{cases}$

La funci√≥n `factorial` utiliza una estructura de control llamada `if... else` para verificar si el valor de $n$ es igual a $0$. Si ese es el caso, la funci√≥n devuelve $1$ porque el factorial de $0=1$. De lo contrario, la funci√≥n utiliza la recursividad para calcular el factorial de $n - 1$ y luego lo multiplica por $n$.

El c√≥digo tambi√©n utiliza un bucle `while` para verificar si el valor de $n$ es negativo. Si es as√≠, se muestra un mensaje de error y se solicita al usuario que ingrese un valor positivo.

Finalmente, el c√≥digo llama a la funci√≥n `factorial` con el valor de $n$ ingresado por el usuario y asigna el resultado a la variable `fact`. Luego, se muestra el resultado del c√°lculo del factorial utilizando una cadena de formato que incluye el valor de $n$ y su factorial correspondiente.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    
print("Calcula el factorial de n")
n = int(input("Ingrese el factorial a calcular: "))

while n < 0:
    print("El valor de n no puede ser negativo")
    n = int(input("Ingrese el factorial a calcular: "))
    
fact = factorial(n)
print(f"El factorial de {n} es: {fact}")
```

## 2. Fibonacci ‚ú®

> **Datos de entrada**: `n`, t√©rmino a calcular de la secuencia de Fibonacci, con $n\geq0$.  
**Datos de salida**: `fibo_term`, t√©rmino `n` de la secuencia de Fibonacci.

A continuaci√≥n, el c√≥digo define la funci√≥n `fibo` que calcula el n-√©simo t√©rmino de la serie de Fibonacci. La serie de Fibonacci es una secuencia de n√∫meros en la que cada n√∫mero es la suma de los dos n√∫meros anteriores.  
Es decir, tenemos: $a_{n}=\begin{cases}
0 & \text{ si } n=0 \\ 
1 & \text{ si } n=1 \\ 
1 & \text{ si } n=2 \\ 
a_{n-1}+a_{n-2} & \text{ si } n>2 
\end{cases}$

La funci√≥n `fibo` utiliza una estructura de control llamada `if... else` para verificar si el valor de $n$ es menor o igual a $1$. Si ese es el caso, la funci√≥n devuelve $n$, ya que el primer y segundo t√©rmino de la serie de Fibonacci son $0$ y $1$, respectivamente.

El c√≥digo tambi√©n utiliza un bucle `while` para verificar si el valor de $n$ es negativo. Si es as√≠, se muestra un mensaje de error y se solicita al usuario que ingrese un valor positivo.

Finalmente, el c√≥digo llama a la funci√≥n `fibo` con el valor de $n$ ingresado por el usuario y asigna el resultado a la variable `fibo_term`. Luego, se muestra el resultado del c√°lculo del n-√©simo t√©rmino de la serie de Fibonacci utilizando una cadena de formato que incluye el valor de $n$ y su correspondiente t√©rmino en la serie.

```python
def fibo(n):
    if n <= 1:
        return n
    else: 
        return fibo(n - 1) + fibo(n - 2)
    
print("Calcula el n-√©simo t√©rmino de Fibonacci")

n = int(input("Ingrese el n-√©simo termino a calcular: "))

while n < 0:
    print("El valor de n no puede ser negativo")
    n = int(input("Ingrese el t√©rmino a calcular: "))

fibo_term = fibo(n)
print(f"El t√©rmino a_{n} = {fibo_term}")
```

## 3. Torres de Hanoi ‚ú®

> **Datos de entrada**: `n`, n√∫mero de discos que el usuario desea mover, con $n\geq0$; `a`, que representa la torre inicial, `b`, la torre auxiliar, y `c` la torre de destino.   
**Datos de salida**: `mov`, n√∫mero de movimientos para mover `n` discos de la torre inicial al objetivo.

El siguiente c√≥digo resuelve el problema de la torre de Hanoi, que consiste en mover una pila de discos de una torre a otra, utilizando una tercera torre como espacio temporal. El objetivo es mover la pila de discos de una torre inicial a una torre objetivo, siguiendo una serie de reglas que indican que solo se puede mover un disco a la vez, y que un disco m√°s grande no puede estar sobre uno m√°s peque√±o.

El problema de las torres de hanoi recursivamente es: 
$a_{n}=\begin{cases}
0 & \text{ si } n=0 \\ 
1 & \text{ si } n=1 \\ 
2a_{n-1}+1 & \text{ si } n>1 
\end{cases}$

La funci√≥n `hanoi` implementa la soluci√≥n recursiva para el problema de la torre de Hanoi, tomando en cuenta que se cuenta con 3 torres. Si el n√∫mero de discos es igual a $1$, la funci√≥n imprime un mensaje que indica c√≥mo mover el disco desde la torre $a$ a la torre $b$, y devuelve $1$. De lo contrario, la funci√≥n se llama a s√≠ misma tres veces con diferentes par√°metros para mover los discos. La primera llamada mueve los $n-1$ discos m√°s peque√±os de la torre $a$ a la torre temporal, utilizando la torre $b$ como espacio temporal. La segunda llamada mueve el disco m√°s grande de la torre $a$ a la torre $b$. La tercera llamada mueve los $n-1$ discos m√°s peque√±os de la torre temporal a la torre $b$, utilizando la torre $a$ como espacio temporal. La funci√≥n devuelve el n√∫mero total de movimientos realizados para mover los discos.

El c√≥digo tambi√©n utiliza un bucle `while` para verificar si el valor de $n$ es negativo. Si es as√≠, se muestra un mensaje de error y se solicita al usuario que ingrese un valor positivo.

Finalmente, el c√≥digo llama a la funci√≥n `hanoi` con los par√°metros necesarios para mover los `n` discos ingresados por el usuario, y muestra el n√∫mero total de movimientos necesarios para mover los discos utilizando una cadena de formato.

```python
def hanoi(a, temp, b, n):
    if n == 1:
        print(f"Mueve disco de torre {a}, a torre {b}")
        return 1
    else:
        mov = hanoi(a, b, temp, n-1)
        mov += hanoi(a, temp, b, 1)
        mov += hanoi(temp, a, b, n-1)
        
        return mov
    
print("Calcula n√∫mero de movimientos para mover n discos")

n = int(input("Ingrese el n¬∞ de discos: "))

while n < 0:
    print("El valor de n no puede ser negativo")
    n = int(input("Ingrese el n¬∞ de discos: "))

    
if n == 0:
    print(f"Se necesitan 0 movimientos para mover {n} discos")
else:
    mov = hanoi(1, 2, 3, n)
    print(f"Se necesitan {mov} movimientos para mover {n} discos")
```

## 4. Multiplicaci√≥n Entera ‚ú®

> **Datos de entrada**: `a` y `b`, son los n√∫meros a multiplicar, con $a,b\geq0$.  
**Datos de salida**: `prod`, es el producto de $a\times b$.

El siguiente c√≥digo implementa la funci√≥n `multi_entera` que toma dos n√∫meros enteros $a$ y $b$ como entrada y devuelve su multiplicaci√≥n como salida. La multiplicaci√≥n se realiza mediante un enfoque de recursi√≥n.

En la funci√≥n, se comprueba si $b=0$. Si es as√≠, la funci√≥n devuelve cero, ya que cualquier n√∫mero multiplicado por cero es cero. Si $b \neq 0$, la funci√≥n devuelve la suma de $a$ y la llamada a s√≠ misma con los argumentos `a` y `b-1`. Esta llamada a s√≠ misma se realiza de forma recursiva hasta que `b` sea cero, en cuyo caso la funci√≥n devuelve cero.
Recursivamente, entonces: $a\times b=\begin{cases}
a + (a\times (b-1)) & \text{ si } b>0 \\ 
0 & \text{ si } b=0 \\ 
\end{cases}$

Despu√©s de definir la funci√≥n `multi_entera`, se solicita al usuario que ingrese dos n√∫meros enteros $a$ y $b$ y se validan que ambos n√∫meros no sean negativos.

Finalmente, se llama a la funci√≥n `multi_entera` con los valores de $a$ y $b$ ingresados por el usuario y se imprime el resultado de la multiplicaci√≥n en la pantalla.

```python
def multi_entera(a, b):
    if b == 0:
        return 0
    else:
        return a + multi_entera(a, b-1)

print("Calcule la multiplicaci√≥n de dos enteros, a * b")

a = int(input("Ingrese a: "))
while a < 0:
    print("a no puede ser un n√∫mero negativo")
    a = int(input("Ingrese a: "))
b = int(input("Ingrese b: "))
while b < 0:
    print("b no puede ser un n√∫mero negativo")
    b = int(input("Ingrese b: "))

prod = multi_entera(a, b)
print(f"El resultado de multiplicar {a} por {b} es: {prod}")
```

## 5. Exponenciaci√≥n Entera ‚ú®

> **Datos de entrada**: `a`, como la base, y `b`, como el exponente, con $a,b\geq0$. En este caso, `a` y `b` no pueden ser 0 simult√°neamente, es decir: $(a=0 \wedge b>0 )\vee (a>0 \wedge b = 0 )$.  
**Datos de salida**: `elev`, el resultado de elevar $a^b$.

A continuaci√≥n, se implementa la funci√≥n `expo`, que calcula el resultado de elevar un n√∫mero entero a una potencia entera. La funci√≥n toma dos argumentos: `a` y `b`. La funci√≥n utiliza una t√©cnica de recursi√≥n para calcular la exponenciaci√≥n.

La definici√≥n recursiva de esta operaci√≥n, se deriva de la definici√≥n de la potencia como una multiplicaci√≥n abreviada, es decir: $a^b=a\times a^{b-1}$

Sin embargo, se pueden buscar otras formas de definir recursivamente la operaci√≥n propuesta, y obtener beneficios en t√©rminos de eficiencia del algoritmo. Consider√°ndose este hecho, puede definirse recursivamente la operaci√≥n potencia como: $a^b = \begin{cases}
a^{b/2} \times a^{b/2} & \text{ si } b>0 \wedge b~es~par \\ 
a \times a^{b-1} & \text{ si } b>0 \wedge b~es~impar \\ 
1 & \text{ si } b=0 
\end{cases}$

Despu√©s de definir la funci√≥n `expo`, el c√≥digo le pide al usuario que ingrese dos n√∫meros enteros $a$ y $b$. Si $a$ o $b$ es menor que cero, se le pide al usuario que vuelva a ingresar el n√∫mero hasta que ingrese un n√∫mero v√°lido. Si tanto $a$ como $b$ son cero, la funci√≥n imprime un mensaje indicando que el resultado est√° indefinido. De lo contrario, el c√≥digo llama a la funci√≥n `expo` con los valores de `a` y `b` ingresados y muestra el resultado de la exponenciaci√≥n.

```python
def expo(a, b):
    if b == 0:
        return 1
    elif b % 2 == 0:
        temp = expo(a, b/2)
        return temp * temp
    else:
        return a * expo(a, b - 1)

print("Calcule la exponenciaci√≥n de dos enteros, a^b")

a = int(input("Ingrese a: "))
while a < 0:
    print("a no puede ser un n√∫mero negativo")
    a = int(input("Ingrese a: "))
b = int(input("Ingrese b: "))
while b < 0:
    print("b no puede ser un n√∫mero negativo")
    b = int(input("Ingrese b: "))

if a == 0 and b == 0:
    print(f"El resultado de elevar {a} a la {b} est√° indefinido")  
else:
    elev = expo(a, b)
    print(f"El resultado de elevar {a} a la {b} potencia es: {elev}")    
```

## 6. M√°ximo Com√∫n Divisor ‚ú®

> **Datos de entrada**: `a` y `b`, de los cuales calcularemos su MCD. Igualmente, `a` y `b` no pueden ser 0 simultaneamente, es decir: $(a=0 \wedge b>0 )\vee (a>0 \wedge b = 0 )$  
**Datos de salida**: `maxcd`, el MCD entre `a` y `b`.

El siguiente c√≥digo, es una implementaci√≥n del algoritmo de Euclides para calcular el MCD de dos n√∫meros enteros. La funci√≥n se llama `mcd` y toma dos argumentos: `a` y `b`.

En este caso, la funci√≥n utiliza una t√©cnica de recursi√≥n, pero se basa en la identidad de Euclides para calcular el MCD. La identidad de Euclides establece que el MCD de dos n√∫meros $a$ y $b$ es igual al MCD de $b$ y el resto de $a$ dividido por $b$. 

Es decir: 
$mcd(a,b) = \begin{cases}
b & \text{ si } a=0 \\ 
a & \text{ si } b=0 \\ 
mcd(b, a~\%~b) & \text{ si } (a=0 \wedge b>0 )\vee (a>0 \wedge b = 0)
\end{cases}$

Por lo tanto, la funci√≥n comienza con las siguientes condiciones: si $b=0$, la funci√≥n devuelve el $a$, y si $a=0$, la funci√≥n devuelve el $b$. Si ambos n√∫meros son iguales a cero, la funci√≥n indica que el MCD no est√° definido. Si ninguna de las condiciones base anteriores se cumple, la funci√≥n llama a s√≠ misma con los argumentos `b` y `a % b`, que es el resto de la divisi√≥n de $a$ entre $b$. Esta llamada recursiva contin√∫a hasta que se alcanza una de las condiciones base e imprime `maxcd` como el MCD entre $a$ y $b$

```python
def mcd(a, b):
    if b == 0:
        return a
    elif a == 0:
        return b
    else:
        return mcd(b, a % b)

print("Calcule el m√°ximo comun divisor de a y b")

a = int(input("Ingrese a: "))
while a < 0:
    print("a no puede ser un n√∫mero negativo")
    a = int(input("Ingrese a: "))
b = int(input("Ingrese b: "))
while b < 0:
    print("b no puede ser un n√∫mero negativo")
    b = int(input("Ingrese b: "))

if a == 0 and b == 0:
    print(f"El m√°ximo comun divisor de 0 y 0 no est√° definido")  
else:
    maxcd = mcd(a, b)
    print(f"El MCD de {a} y {b} es {maxcd}")    
```

## 7. N√∫mero Mayor ‚ú®

> **Datos de entrada**: `nums`, n√∫meros que el usuario va ingresando separados por un espacio, los n√∫meros son reales.  
**Datos de salida**: `max_num`, el n√∫mero mayor de entre la lista de n√∫meros ingresada.

Este c√≥digo define una funci√≥n `max_fun` que encuentra el n√∫mero m√°ximo en una lista de n√∫meros `nums` utilizando la t√©cnica de recursi√≥n.

Si la lista tiene m√°s de un n√∫mero, la funci√≥n extrae el √∫ltimo n√∫mero de la lista utilizando el m√©todo `pop()` de Python y lo almacena en la variable `numero`. Luego, la funci√≥n llama a s√≠ misma con la lista restante y almacena el valor devuelto en la variable `max_act`, que representa el n√∫mero que es mayor por el momento.

La funci√≥n compara `max_act` con `numero`. Si `max_act` es mayor que `numero`, devuelve `max_act`. De lo contrario, devuelve `numero`.

Luego, el programa principal solicita al usuario que ingrese una lista de n√∫meros separados por espacios y crea una lista de n√∫meros a partir de ella, ya que Python interpretar√≠a lo ingresado por el usuario como `string`. Si la lista est√° vac√≠a, el programa solicita al usuario que ingrese una lista de n√∫meros nuevamente hasta que se ingrese al menos un n√∫mero.

Finalmente, el programa llama a la funci√≥n `max_fun` con la lista de n√∫meros y muestra el n√∫mero m√°ximo en la lista.

```python
def max_fun(nums):
    if len(nums) == 1:
        return nums[0]
    else:
        numero = float(nums.pop())
        max_act = max_fun(nums)
        if max_act > numero:
            return max_act
        else:
            return numero

print("Calcule el n√∫mero m√°ximo de un conjunto de n√∫meros")

nums = input("Ingrese una lista de n√∫meros separados por un espacio: ")
num_a_lista = [float(n) for n in nums.split()]

while len(num_a_lista) == 0:
    nums = input("No ha ingresado ningun n√∫mero. Ingrese una lista de n√∫meros separados por un espacio: ")
    num_a_lista = [float(n) for n in nums.split()]

max_num = max_fun(num_a_lista)

print(f"El n√∫mero m√°ximo de la lista es: {max_num}")
```

## 9. N√∫mero Combinatorio ‚ú®

> **Datos de entrada**: `n`,es el n√∫mero total de elementos, y `k` el n√∫mero de elementos a seleccinar.  
**Datos de salida**: `result`, el n√∫mero de combinaciones totales de `n` en `k`.

Este c√≥digo es una implementaci√≥n de la f√≥rmula matem√°tica para calcular el n√∫mero combinatorio de $n$ en $k$, tambi√©n conocido como coeficiente binomial.

Para elegir $k$ elementos de un conjunto de $n$ elementos, podemos tomar un elemento espec√≠fico como parte de esos $k$ elementos o no. Si tomamos un elemento espec√≠fico como parte de los $k$ elementos, entonces debemos elegir $k-1$ elementos de los $n-1$ elementos restantes. Si no tomamos ese elemento espec√≠fico como parte de los $k$ elementos, entonces debemos elegir $k$ elementos de los $n-1$ elementos restantes.

Al sumar estas dos opciones, obtenemos el n√∫mero total de formas de elegir $k$ elementos de un conjunto de $n$ elementos. Y debido a que la f√≥rmula se define recursivamente, podemos utilizarla para calcular el n√∫mero combinatorio de cualquier par de valores $n$ y $k$. 

Recursivamente, esto es: $C_{n, k}=C_{n-1,~k-1}+C_{n-1,~k}$

La funci√≥n definida al inicio, `combinaciones`, utiliza recursi√≥n para calcular el n√∫mero combinatorio. Si $k=0 \vee k=n$, entonces retorna $1$, porque la f√≥rmula combinatoria de un conjunto vac√≠o o del conjunto completo siempre es $1$. De lo contrario, utiliza la f√≥rmula recursiva mencionada anteriormente.

El programa luego solicita al usuario que ingrese los valores de `n` y `k`, y los valida para asegurarse de que no sean n√∫meros negativos. Si $k$ es mayor que $n$, significa que el n√∫mero combinatorio no est√° definido y se imprime un mensaje que indica esto. Si $k$ es menor o igual que $n$, el programa utiliza la funci√≥n `combinaciones` para calcular el n√∫mero combinatorio y lo imprime en pantalla.

```python
def combinaciones(n, k):
    if k == 0 or k == n:
        return 1
    else:
        return combinaciones(n-1, k-1) + combinaciones(n-1, k)
    
print("Calcule el n√∫mero combinatorio de n en k, enteros")

n = int(input("Ingrese n: "))
while n < 0:
    print("n no puede ser un n√∫mero negativo")
    n = int(input("Ingrese n: "))
k = int(input("Ingrese k: "))
while k < 0:
    print("k no puede ser un n√∫mero negativo")
    k = int(input("Ingrese k: "))

if k > n:
    print(f"El n√∫mero combinatorio de {n} en {k} no est√° definido")  
else:
    result = combinaciones(n, k)
    print(f"El n√∫mero combinatorio de {n} en {k} es {result}")  
```

## 10. Reversa de una Lista ‚ú®

> **Datos de entrada**: `nums`,es la de n√∫meros ingresada por el usuario, estos pueden ser reales.  
**Datos de salida**: `lista_rev`, es la lista original al rev√©s.

Este programa calcula la reversa de una lista de n√∫meros ingresada por el usuario.

Primero, se define una funci√≥n `reversa` que toma una lista `nums` como argumento y utiliza recursi√≥n para calcular su reversa. Si la longitud de la lista es igual a $1$, entonces se retorna la misma lista porque una lista de un solo elemento es igual a su reversa. De lo contrario, se utiliza la f√≥rmula recursiva: la reversa de la lista `nums` es igual a la concatenaci√≥n del √∫ltimo elemento de `nums` con la reversa del resto de `nums`, es decir, `nums[:-1]`.

Despu√©s, el programa solicita al usuario que ingrese una lista de n√∫meros separados por un espacio, y la convierte en una lista de enteros. Si la lista est√° vac√≠a, se le pide al usuario que ingrese una lista v√°lida hasta que lo haga.

Luego, se llama a la funci√≥n reversa con la lista de n√∫meros ingresada por el usuario y se almacena el resultado en una variable `lista_rev`. Finalmente, se imprime en pantalla la lista original y su reversa.

```python
def reversa(nums):
    if len(nums) == 1:
        return nums
    else:
        return [nums[-1]] + reversa(nums[:-1])

print("Calcule la reversa de una lista")

nums = input("Ingrese una lista de n√∫meros separados por un espacio: ")
num_a_lista = [float(n) for n in nums.split()]

while len(num_a_lista) == 0:
    nums = input("No ha ingresado ningun n√∫mero. Ingrese una lista de n√∫meros separados por un espacio: ")
    num_a_lista = [float(n) for n in nums.split()]

lista_rev = reversa(num_a_lista)

print(f"La lista {num_a_lista} al rev√©s es {lista_rev}")
```
